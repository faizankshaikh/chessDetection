```python
# define data dir
data_dir = Path('/content/drive/MyDrive/MAID/502 CI/project/data')
```

```python
# load training metadata
train_data = pd.read_csv(data_dir / 'train' / 'annot_csv' / 'train.csv')
train_data.head()
```

```python
# load validation metadata
valid_data = pd.read_csv(data_dir / 'valid' / 'annot_csv' / 'valid.csv')
valid_data.head()
```

```python
%%bash

mkdir -p "/content/drive/MyDrive/MAID/502 CI/project/data"/{train/labels/{blackKing,whiteKing,blackQueen,whiteQueen},valid/labels/{blackKing,whiteKing,blackQueen,whiteQueen}}
```

```python
# create training data
for idx, row in train_data.iterrows():

    # get metadata
    img_name = row.iloc[0]
    class_name = row.iloc[1]
    xcenter = row.iloc[2]
    ycenter = row.iloc[3]
    width = row.iloc[4]
    height = row.iloc[5]

    # extract label
    _stem = Path(img_name).stem
    label = _stem.split('_')[1][0]

    # define annotation path
    annot_name = _stem + '.txt'
    annot_path = str(data_dir / 'train' / 'labels' / class_name / annot_name)

    # normalize data for yolo
    norm_xcenter = xcenter / 224
    norm_ycenter = ycenter / 224
    norm_width = width / 224
    norm_height = height / 224

    # save annotation
    with open(annot_path, 'w') as f:
        # pattern is
        #   label normalized_xcenter normalized_ycenter normalized_width normalized_height
        f.write(label + ' ' + str(norm_xcenter) + ' ' + str(norm_ycenter) + ' ' + str(norm_width) + ' ' + str(norm_height))
```

```python
# create validation data
for idx, row in valid_data.iterrows():

    # get metadata
    img_name = row.iloc[0]
    class_name = row.iloc[1]
    xcenter = row.iloc[2]
    ycenter = row.iloc[3]
    width = row.iloc[4]
    height = row.iloc[5]

    # extract label
    _stem = Path(img_name).stem
    label = _stem.split('_')[1][0]

    # define annotation path
    annot_name = _stem + '.txt'
    annot_path = str(data_dir / 'valid' / 'labels' / class_name / annot_name)

    # normalize data for yolo
    norm_xcenter = xcenter / 224
    norm_ycenter = ycenter / 224
    norm_width = width / 224
    norm_height = height / 224

    # save annotation
    with open(annot_path, 'w') as f:
        # pattern is
        #   label normalized_xcenter normalized_ycenter normalized_width normalized_height
        f.write(label + ' ' + str(norm_xcenter) + ' ' + str(norm_ycenter) + ' ' + str(norm_width) + ' ' + str(norm_height))
```
